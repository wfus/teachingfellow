```python
import networkx
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
from tqdm.notebook import tqdm
import pandas as pd
```

```python
def plot_degree(net):
    degrees = [net.degree[a] for a in net.nodes]
    sns.distplot(degrees)
    plt.ylabel("Density")
    plt.xlabel("Degree of Researcher")
    plt.xlim(0, 15)
    plt.show()
```

```python
net = networkx.read_gml('netscience.gml')
```

```python
plot_degree(net)
```

```python
reduced_net = net.copy()
node_degs = [(node, net.degree(node)) for node in reduced_net.nodes]
node_degs = sorted(node_degs, key=lambda x: -x[-1])
top10percent_nodes = node_degs[:len(node_degs) // 10]
for node, _ in top10percent_nodes:
    reduced_net.remove_node(node)
```

```python
plot_degree(reduced_net)
```

```python

def simulate_virality(net, p, sim='dfs'):
    """Simulate virality of our network - assuming people that have already shared the video
    will not reshare it."""
    visited = set()
    # Select 10 random people to spread it first.
    people = list(net.nodes)  
    queue = np.random.choice(people, 10).tolist()
    while queue:
        infected = queue.pop()
        if infected not in visited:
            visited.add(infected)
            for neighbor in net.neighbors(infected):
                if np.random.random() < p:
                    if sim == 'dfs':
                        queue.append(neighbor)
                    elif sim == 'bfs':
                        queue = [neighbor] + queue
                    else:
                        raise RuntimeError("Invalid sim type")
                        
    return len(visited) / len(people)
```

```python
NUM_TRIALS = 100
data = []

for transmission_prob in tqdm(np.linspace(0, 1, 101)):
    for sim_type, net_type in [('regular', net), ('top10percent_removed', reduced_net)]:
        for trial in range(NUM_TRIALS):
            data.append((transmission_prob, trial, sim_type, 
                         simulate_virality(net_type, transmission_prob, sim='dfs')))
        
df = pd.DataFrame(data, columns=['p', 'trial', 'sim_type', 'virality'])
```

```python
plt.figure(figsize=(10, 4))
sns.lineplot(y='virality', x='p', hue='sim_type', data=df)
plt.show()
```
